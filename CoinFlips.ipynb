```python
import numpy.random as random
import numpy as np 
import matplotlib.pyplot as plt 

#k coins, n times
def generate_counts(k=1000, n=100): #(k=1000, n=100): 
    X=2*(random.rand(k,n)>0.5)-1
    S = np.sum(X, axis=0)
    return S
```

```python
suma = generate_counts(k=1, n=100000)
print(np.sum(suma))
```

```python
plt.figure(figsize=(10,4))
plt.hist(generate_counts(n=1000), bins=9)
plt.xticks(np.arange(-1000,1250,250))
plt.xlabel('Sum')
plt.ylabel('Count')
plt.title('Histogram of coin flip sum when flipping a fair coin 1000 times')
plt.grid()
plt.show()
```
