```python
import re
from bs4 import BeautifulSoup
from urllib.request import FancyURLopener

class MyOpener(FancyURLopener):
#se utiliza lo sigte para autenticarme/saltar la auth del servidor del site q se va a scrapping
    version = 'Mozilla/5.0 (Windows; U; Windows NT 5.1; it; rv:1.8.1.11) Gecko/20071127 Firefox/2.0.0.11'
myopener = MyOpener()

news_content = []

for i in links_news:

    #print (i)
    link = i

    html = myopener.open(i).read()
    soup = BeautifulSoup(html,"lxml") #beautiful soup solo necesita url
    
    parrafo = ''
     
    for div in soup.findAll('div', {'class': 'news-text-content'}):
        for b in soup.findAll('p'):
            parrafo = parrafo +' '+ b.text +' '
            #print (b.text)
    news_content.append(parrafo)
```

```python
j=1

for item in news_content:
    #guardar a un txt
    filename_str = filename+str(j)+'.txt'
    links_list = open(filename_str, 'w') 
    links_list.write("%s\n" % item)
    j=j+1

links_list.close()
```
